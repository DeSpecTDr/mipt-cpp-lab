```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
datadir = Path("mipt/lab01/data")
```

```python
def plot(name, title=None, exp=False):
    path = datadir / name
    data = np.loadtxt(path)
    plt.title(name if title is None else title)
    plt.xlabel("N")
    plt.ylabel("T, ms")
    x = data[:, 0]
    y = data[:, 1]
    # if exp:
    #     x = np.exp(x)
    plt.plot(x, y, label=path.stem)
```

```python
plot("lin.txt")
plot("linWorst.txt", "$O(n)\\ linear\\ search$")
plt.legend()
```

```python
plot("bin.txt")
plot("binWorst.txt", "$O(log n)\\ binary\\ search$")
plt.legend()
```

```python
plot("bin.txt")
plot("binWorst.txt", "$O(log n)\\ binary\\ search\\ (log\\ scale)$")
plt.legend()
plt.xscale("log")
```

```python
plot("n2sum.txt", "$O(n^2)\\ sum\\ search\\ (average)$")
```

```python
plot("nsum.txt", "$O(n)\\ sum\\ search\\ (average)$")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)
names = ["random", "fixed", "correlated"]
abc = ["A", "B", "C"]
for i, ax in enumerate(axs):
    for j in range(3):
        name = f"find{abc[j]}{names[i]}.txt"
        path = datadir / name
        data = np.loadtxt(path)
        x = data[:, 0]
        y = data[:, 1]
        ax.plot(x, y, label=path.stem)
        ax.legend()
        ax.set_xlabel("N")
        ax.set_ylabel("T, ms")
```
