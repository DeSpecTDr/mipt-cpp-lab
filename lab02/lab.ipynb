```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(file, name, label=None, linearize=False):
    x = np.loadtxt(file)
    if linearize:
        plt.plot(x[:, 0], np.sqrt(x[:, 1]), label=label)
    else:
        plt.plot(x[:, 0], x[:, 1], label=label)
    plt.title(name)
```

# Vector

```python
plot("d1.txt", "size = size + 1")
plt.xlabel("n")
plt.ylabel("$T, мс$")
```

```python
plot("d1.txt", "size = size + 1", linearize=True)
plt.xlabel("n")
plt.ylabel("$\\sqrt{{T}}, мс$")
```

```python
plot("d100.txt", "size = size + 100")
plt.xlabel("n")
plt.ylabel("$T, мс$")
```

```python
plot("d100.txt", "size = size + 100", linearize=True)
plt.xlabel("n")
plt.ylabel("$\\sqrt{{T}}, мс$")
```
